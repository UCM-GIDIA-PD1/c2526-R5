```python
# Valores únicos en la columna de referencia (DF2)
referencia = set(df['stop_id'].str[:3])

# Valores en DF1 que son NaN o que simplemente no están en la referencia
faltantes_en_df2 = df2[~df2['GTFS Stop ID'].isin(referencia)]

print(f"Total de registros en DF2 sin coincidencia en DF1: {len(faltantes_en_df2)}")
```

```python
# Identifica cuáles son los IDs que sobran en tu dataset de entrenamiento
ids_entrenamiento = set(df['stop_id'].str[:3].unique())
ids_catalogo = set(df2['GTFS Stop ID'].str[:3].unique())

sobrantes_en_entrenamiento = ids_entrenamiento - ids_catalogo

print(f"IDs en el tráfico que NO están en el catálogo: {sobrantes_en_entrenamiento}")
```
